```python
import random
import pandas as pd

a = [] #this is one variable, between 20 and 60
b = [] # this is another variable, between 1 and 900
c = [] # this is another variable, between 0 and 1
d = []
e = []
f = []
g = []
h = []
j = []
k = []
m = []
p_ups = []
d_offs = []
team1 = [
    "Automation", "Application", "Global", "International", "Operations", "Inventory", "Capital"
]
team2 = [
    "Initiative", "Implementation", "Investment", "Management", "Reporting", "Market", "Implementation", "Communication"
]
team3 = [
    "Process", "Team", "Office", "Group", "Service", "Product"
]
l = {} # this will hold my dataframe. 
coords = pd.read_csv('src/locations.csv')
for index, row in coords.iterrows(): 
    p_ups.append([row['pickup_latitude'], row['pickup_longitude']])
    d_offs.append([row['dropoff_latitude'], row['dropoff_longitude']])
for i in range(0, 9700):
    a.append(random.randint(20, 60))
    b.append(random.randint(1, 900))
    c.append(random.randint(0, 1))
    d.append(random.randint(0, 100))
    e.append(random.randint(0, 100))
    f.append(random.randint(1, 44) * 12.112)
    g.append(random.randint(2, 90) / 2)
    h.append(random.randint(1, 20000) / 20)
    j.append(p_ups[random.randint(0, 300)])
    k.append(d_offs[random.randint(0, 300)])
    mean = team1[random.randint(0, 6)] + ' ' + team2[random.randint(0, 7)] + ' ' + team3[random.randint(0, 5)]
    m.append(mean)

l['Process']  = m  
l['Value'] = a
l['Frequency'] = b
l['Accepted'] = c
l['Adaptation_Rate'] = d
l['Inclusion_Rate'] = e
l['Cost'] = f
l['Rides'] = g
l['Purchased'] = h
l['Starting_Position'] = j
l['Ending_Position'] = k

df = pd.DataFrame(l)
```

```python
print(df)
writer = pd.ExcelWriter('fake_data.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Sheet1')
writer.save()

```
